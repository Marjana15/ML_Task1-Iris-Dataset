```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris-flower-dataset.csv')
df.head()
```

```python
# Mapping species names to numerical values
species_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df['target'] = df['species'].map(species_mapping)
df.head()
```

```python
df[df['target'] == 1].head()  # Filter for Iris-versicolor
df[df['target'] == 2].head()  # Filter for Iris-virginica
```

```python
# Step 7: Plotting a pairplot for visualization
sns.pairplot(df, hue='species')
plt.show()

# Step 8: Summary statistics for each species
summary_stats = df.groupby('species').describe()
print(summary_stats)
```

```python
df_slice = df[45:55]
df_slice
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal_length'], df0['sepal_width'], color="green", marker='+')
plt.scatter(df1['sepal_length'], df1['sepal_width'], color="cyan", marker='.')
plt.scatter(df2['sepal_length'], df2['sepal_width'], color="magenta", marker='*')
plt.show()
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal_length'], df0['petal_width'], color="orange", marker='+')
plt.scatter(df1['petal_length'], df1['petal_width'], color="red", marker='.')
plt.show()
```

```python
df_bar = pd.DataFrame({
    'Sepal Length': [5.1, 5.9, 6.3],
    'Sepal Width': [3.5, 3.0, 3.3],
    'Petal Length': [1.4, 4.2, 6.0],
    'Petal Width': [0.2, 1.5, 2.5],
}, index=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])

df_bar.plot(kind='bar', figsize=(10, 6), color=["#FF9999", "#99CCFF", "#99FF99", "#FF00FF"])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))

# Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', hue='species', data=df, palette={'Iris-setosa': 'lightblue', 'Iris-versicolor': 'lightgreen', 'Iris-virginica': 'lightcoral'})
plt.title('Sepal Length')

# Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', hue='species', data=df, palette={'Iris-setosa': 'lightblue', 'Iris-versicolor': 'lightgreen', 'Iris-virginica': 'lightcoral'})
plt.title('Sepal Width')

# Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', hue='species', data=df, palette={'Iris-setosa': 'lightblue', 'Iris-versicolor': 'lightgreen', 'Iris-virginica': 'lightcoral'})
plt.title('Petal Length')

# Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', hue='species', data=df, palette={'Iris-setosa': 'lightblue', 'Iris-versicolor': 'lightgreen', 'Iris-virginica': 'lightcoral'})
plt.title('Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Replace numeric species with actual names if not already done
species_mapping = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
df['species'] = df['target'].map(species_mapping)

# Create the scatter plot with regression lines
sns.lmplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette=['orange', 'green', 'purple'], markers=['o', 's', 'D'], height=4.5, aspect=1.6)
plt.show()
```
